## Q3: Identify periods with the strongest and weakest retention rates. What strategies could be implemented to improve retention during weaker months?


````python
#Retention formula
# Retention (Month X) = customers in X who return in X+1 / customers in X
````

````python
online_sales[['CustomerID', 'Transaction_month']].head()
````

````python
online_sales[['CustomerID', 'Transaction_month']]
````

````python
online_sales.head()
````

````python
#Creating a Retention base table where we will only have unique CustomerID with Transaction_month no duplicate
retention_df = online_sales[['CustomerID', 'Transaction_month']].drop_duplicates()
retention_df.head()
````

````python
#Build customer sets per month(creating set of customer purchased in each month)
monthly_customers =(retention_df.groupby('Transaction_month')['CustomerID'].apply(set))
monthly_customers.head()
````

````python
#Calculate retention month by month

retention_results = []

for month in range(1,12):
  current_month_customers = monthly_customers.get(month, set())
  next_month_customers = monthly_customers.get(month +1, set())

  if len(current_month_customers)== 0:
    retention_rate = 0
  else:
    retained =current_month_customers & next_month_customers
    retention_rate = len(retained) / len(current_month_customers)

  retention_results.append({'month':month,
                            'retention_rate':retention_rate})
````

````python
#Convert to Dataframe
retention_rate_df = pd.DataFrame(retention_results)
retention_rate_df
````

````python
from matplotlib import pyplot as plt
retention_rate_df['retention_rate'].plot(kind='line', figsize=(8, 4), title='retention_rate')
plt.gca().spines[['top', 'right']].set_visible(False)
````

````python
# Identify Strongest & weakest retention months
best_month = retention_rate_df.loc[retention_rate_df['retention_rate'].idxmax()]
worst_month = retention_rate_df.loc[retention_rate_df['retention_rate'].idxmin()]

best_month, worst_month
````

#### i) Observations


*   Retention rates show meaningful month-to-month variation, indicating changing customer engagement levels.
*   Certain months show strong customer continuity into the following month.
*   Some months experience sharp drop-offs, indicating higher churn.
---

#### ii) Business Interpretation


*   Strong retention months may align with:
    - Better post-purchase engagement
    - Ongoing promotions
    - High product satisfaction



*   Weak retention months may indicate:
    - Lack of follow-up communication
    - One-time promotional purchases
    - Reduced perceived value after first purchase
---
#### iii) **Retention Performance Summary**
- July shows the strongest customer retention, indicating higher repeat engagement during this period.
- January exhibits the weakest retention, suggesting higher churn after initial purchases.

---
#### iii) Recommendations


*   For weak retention months:
    - Introduce post-purchase email journeys
    - Offer next-month incentives (credits, coupons)
    - Improve onboarding and product education
    - Run reminder & replenishment campaigns
*   For strong retention months:
    - Study what strategy work (like offers, categories, pricing)
    - Replicate engagement strategies
    - Push loyalty programs when intent is high
---
#### iv) Final Verdict
Retention shows clear month-to-month variation, indicating that customer continuity depends heavily on engagement strategy rather than acquisition volume alone. By strengthening post-purchase touchpoints and targeted re-engagement during weak months, the company can significantly improve overall customer lifetime value.

## Q10: Segment customers into groups such as Premium, Gold, Silver, and Standard. What targeted strategies can be developed for each segment to improve retention and revenue? (Use RFM segmentation techniques)


````python
#Prepare Revenue & Reference Date
sales_discount['Transaction_Date'] = pd.to_datetime(sales_discount['Transaction_Date'])
today = sales_discount['Transaction_Date'].max() + pd.Timedelta(days=1)

today
````

````python
#Calculate RFM Metrics
rfm = sales_discount.groupby('CustomerID').agg({
    'Transaction_Date': lambda x: (today - x.max()).days,  # Recency
    'Transaction_ID': 'count',                             # Frequency
    'Revenue': 'sum'                                       # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm.head()
````

````python
#Assigning R, F, M Scores

rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])

# Frequency & Monetary: higher is better
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm[['R_Score','F_Score','M_Score']] = rfm[['R_Score','F_Score','M_Score']].astype(int)
rfm.head()
````

````python
# Compute RFM Score
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']
rfm.head()
````

````python
#Customer Segmentation
def segment_customer(score):
    if score >= 10:
        return 'Premium'
    elif score >= 8:
        return 'Gold'
    elif score >= 6:
        return 'Silver'
    else:
        return 'Standard'

rfm['Customer_Segment'] = rfm['RFM_Score'].apply(segment_customer)
rfm.head()
````

````python
#Segment Distribution
rfm['Customer_Segment'].value_counts()
````

### i) Observations

* Customers can be clearly segmented into **Premium, Gold, Silver, and Standard** groups based on Recency, Frequency, and Monetary values.
* A smaller proportion of customers contribute a disproportionately large share of total revenue.
* High RFM score customers purchase more frequently, spend more per transaction, and return sooner.
* Lower RFM score customers show infrequent purchases, longer inactivity periods, and lower spend.

---

### ii) Business Interpretation

* **Premium customers** represent the most valuable segment, driving consistent revenue and showing strong brand loyalty.
* **Gold customers** demonstrate high engagement but still have potential to move into the Premium segment.
* **Silver customers** are moderately engaged and are often influenced by pricing, promotions, or convenience.
* **Standard customers** have the highest churn risk and often consist of one-time or low-intent buyers.
* Overall revenue growth is strongly influenced by how well high-value segments are retained while selectively reactivating lower-value customers.

---

### iii) Targeted Strategies by Customer Segment

#### Premium Customers

* Focus on **retention and relationship strengthening** rather than discounts.
* Provide early access to new products, exclusive deals, and loyalty rewards.
* Use personalized recommendations and premium customer support.
* Aim to maximize lifetime value rather than short-term sales.

#### Gold Customers

* Encourage progression into the Premium segment.
* Offer tier-based rewards, limited-time incentives, and loyalty upgrades.
* Use cross-selling and upselling strategies based on purchase history.
* Reinforce brand engagement through personalized communication.

#### Silver Customers

* Target with value-driven promotions and reminders.
* Use bundled offers, seasonal discounts, and re-engagement emails.
* Educate customers about product benefits to increase repeat purchases.
* Focus on increasing purchase frequency gradually.

#### Standard Customers

* Apply cost-efficient reactivation strategies.
* Offer first-return discounts or onboarding incentives.
* Use targeted remarketing instead of heavy discounting.
* Identify low-potential customers to avoid inefficient marketing spend.

---

### iv) Final Verdict

RFM-based customer segmentation highlights that **not all customers contribute equally to revenue or growth**. By aligning marketing, retention, and promotional strategies with customer value tiers, the company can improve retention, optimize marketing spend, and drive sustainable revenue growth. Strategic focus on Premium and Gold segments, combined with selective reactivation of lower-value customers, ensures a balanced and profitable customer portfolio.

## Q12: Group customers by their month of first purchase and analyze retention rates over time. Which cohorts exhibit the highest and lowest retention rates? What strategies can be implemented to improve retention for weaker cohorts?



````python
# Ensure Transaction_Date is datetime
online_sales['Transaction_Date'] = pd.to_datetime(online_sales['Transaction_Date'])

# Create transaction month (year-month for cohort tracking)
online_sales['Transaction_Month'] = online_sales['Transaction_Date'].dt.to_period('M')

# Identify first purchase month for each customer
online_sales['Cohort_Month'] = (
    online_sales
    .groupby('CustomerID')['Transaction_Month']
    .transform('min')
)

online_sales[['CustomerID', 'Transaction_Month', 'Cohort_Month']].head()
````

````python
#Calculate Cohort Index
def cohort_index(df):
    return (df['Transaction_Month'] - df['Cohort_Month']).apply(lambda x: x.n)

online_sales['Cohort_Index'] = cohort_index(online_sales)

online_sales[['CustomerID', 'Cohort_Month', 'Transaction_Month', 'Cohort_Index']].head()
````

````python
#Build Cohort Table
cohort_data = (
    online_sales
    .groupby(['Cohort_Month', 'Cohort_Index'])['CustomerID']
    .nunique()
    .reset_index()
)

cohort_data.head()
````

````python
#Cohort Retention Matrix
cohort_pivot = cohort_data.pivot(
    index='Cohort_Month',
    columns='Cohort_Index',
    values='CustomerID'
)

# Convert to retention rate
cohort_size = cohort_pivot.iloc[:, 0]

cohort_retention = cohort_pivot.divide(cohort_size, axis=0)

cohort_retention
````

````python
#Finding Strongest & Weakest Cohorts
# Average retention per cohort (excluding month 0)
avg_retention = cohort_retention.iloc[:, 1:].mean(axis=1)

highest_retention_cohort = avg_retention.idxmax()
lowest_retention_cohort = avg_retention.idxmin()

highest_retention_cohort, lowest_retention_cohort
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(
    cohort_retention,
    annot=True,
    fmt='.0%',
    cmap='Blues'
)
plt.title('Customer Retention Cohort Analysis')
plt.xlabel('Months Since First Purchase')
plt.ylabel('Cohort Month')
plt.show()
````

### i) Observations

* Customers grouped by their **first purchase month (cohorts)** show varying retention patterns over time.
* Some cohorts retain a higher proportion of customers across multiple months, indicating stronger loyalty.
* Other cohorts experience steep drop-offs after the initial purchase.
* Retention generally declines as the number of months since first purchase increases.
* Clear differences exist between high-performing and low-performing cohorts.

---

### ii) Business Interpretation

* **High-retention cohorts** likely benefited from:

  * Better onboarding experience
  * Relevant promotions or product mix
  * Higher customer intent at acquisition
* **Low-retention cohorts** may reflect:

  * One-time promotional purchases
  * Poor post-purchase engagement
  * Seasonal or low-intent acquisition
